```python
import requests
import pandas as pd
import numpy as np
data = pd.read_csv("data_final.csv")
```

```python
import pandas as pd
from geolite2 import geolite2

def get_country(ip):
    try:
        x = geo.get(ip)
    except ValueError:
        return np.nan
    try:
        return x['country']['names']['en'] if x else np.nan
    except KeyError:
        return np.nan

geo = geolite2.reader()

data['country'] = data['c-ip'].apply(get_country)
geolite2.close()
```

```python
import pandas as pd
from geolite2 import geolite2

def lat(ip):
    try:
        x = geo.get(ip)
    except ValueError:
        return np.nan
    try:
        return x['location']['latitude'] if x else np.nan
    except KeyError:
        return np.nan

geo = geolite2.reader()
data['latitude'] = data['c-ip'].apply(lat)
geolite2.close()
```

```python
import pandas as pd
from geolite2 import geolite2

def long(ip):
    try:
        x = geo.get(ip)
    except ValueError:
        return np.nan
    try:
        return x['postal']['code'] if x else np.nan
    except KeyError:
        return np.nan

geo = geolite2.reader()

data['postal_code'] = data['c-ip'].apply(long)

geolite2.close()
```

```python
import re
def find_robots(a):
    if re.search('robots', a, re.IGNORECASE):
      return ("search_engine")
    else:
      return ("normal_user")

data['type_of_visit'] = data['cs-uri-stem'].apply(find_robots)   
```
